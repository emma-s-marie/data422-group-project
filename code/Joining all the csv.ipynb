```r
library(dplyr)
library(tidyverse)
```

```r
data_crime <- read.csv("Final_crime_by_TA.csv")
data_housing <- read.csv("Final_housedata.csv")
data_population <- read.csv("Final_Population_Data.csv")
data_school <- read.csv("Final_School_Data.csv")
```

```r
# Checking all unique value and count in the house data
unique_values_TA <- unique(data_housing$Territorial_Authority)
unique_values_TA <- sort(unique_values_TA)
unique_values_TA
length(unique_values_TA)
```

```r
# Checking all unique value and count in the population data
unique_values_TA <- unique(data_population$Territorial_Authority)
unique_values_TA <- sort(unique_values_TA)
unique_values_TA
length(unique_values_TA)
```

```r
# Checking all unique value and count in the crime data
unique_values_TA <- unique(data_crime$Territorial_Authority)
unique_values_TA
length(unique_values_TA)
```

```r
# Checking all unique value and count in the school data
unique_values_TA <- unique(data_school$Territorial_Authority)
unique_values_TA
length(unique_values_TA)
```

```r
# joining crime and housing data and find out the count of distinct TA 
crime_housing <- inner_join(data_crime, data_housing, by = "Territorial_Authority")

unique_values_TA <- unique(crime_housing$Territorial_Authority)
unique_values_TA
length(unique_values_TA)
crime_housing
```

```r
# joining crime, housing and population data and find out the count of distinct TA 
crime_housing_population <- inner_join(crime_housing, data_population, by = "Territorial_Authority")
unique_values_TA <- unique(crime_housing_population$Territorial_Authority)
length(unique_values_TA)
crime_housing_population
```

```r
# joining crime, housing, population and school data and find out the count of distinct TA 
all_df <- inner_join(crime_housing_population, data_school, by = "Territorial_Authority")
unique_values_TA <- unique(all_df$Territorial_Authority)
length(unique_values_TA)
all_df
```

```r
all_df %>%
    write_csv("all_df.csv")
```
